**17. What is the difference between a class variable and an instance variable?**

**1. Class Variables:**

Scope: Class variables are defined at the class level, outside of any instance methods. They are shared among all instances of the class.

Access: Class variables can be accessed using the class name or through any instance of the class.

Modification: If a class variable is modified, the change is reflected across all instances of the class.

Usage: Class variables are typically used for data that should be consistent and shared among all instances of the class.

**2. Instance Variables:**

Scope: Instance variables are defined within instance methods, typically within the __init__ method. They are specific to each instance of the class.

Access: Instance variables are accessed using the instance name.

Modification: If an instance variable is modified, the change only affects that particular instance.

Usage: Instance variables are used for data that is unique to each instance of the class.

**20. What is the significance of the super() function in Python?**

The super() function in Python is used to call methods from a parent class (superclass) within a subclass. It is especially useful in the context of inheritance and helps ensure that the methods and properties of the superclass are properly invoked, allowing for better code reusability and maintainability. Here are some key points about the significance of the super() function:

**Access to Parent Class Methods:** The super() function allows you to call methods from the parent class within the subclass, ensuring that the parent class's implementation is properly utilized.

**Avoiding Redundant Code:** By using super(), you can avoid duplicating code that is already defined in the parent class. This promotes code reusability and keeps your codebase DRY (Don't Repeat Yourself).

**Initialization of Parent Class:** In the context of constructors (__init__ methods), super() is used to ensure that the parent class's initialization code is executed, allowing for proper setup of inherited attributes and methods.

**Support for Multiple Inheritance:** When dealing with multiple inheritance, super() ensures that the method resolution order (MRO) is followed, and the appropriate methods from multiple parent classes are called in the correct order.

**22. What is the difference between @staticmethod and @classmethod in Python?**

The @staticmethod and @classmethod decorators in Python are used to define methods within a class that have different behaviors and use cases. Here's a breakdown of the differences between them:

**@staticmethod**

Definition: A staticmethod is a method that belongs to the class but does not have access to the instance (self) or the class itself (cls). It behaves like a regular function defined inside a class.

**@classmethod**

Definition: A classmethod is a method that belongs to the class and has access to the class itself (cls) but not to any instance-specific data. It can modify the class state that applies across all instances.

```python
 5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes
balance and methods to deposit, withdraw, and check balance

class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Deposit amount must be positive")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Invalid withdraw amount or insufficient balance")

    def check_balance(self):
        print(f"Current balance: {self.__balance}")

# Create an instance of BankAccount
account = BankAccount(100)

# Demonstrate deposit, withdraw, and check balance
account.deposit(50)
account.withdraw(30)
account.check_balance()

# Attempt to access the private attribute directly (will raise an AttributeError)
try:
    print(account.__balance)
except AttributeError as e:
    print(e)
```

```python
 16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that
implement the sound() method.

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Cow(Animal):
    def sound(self):
        return "Moo"

class Sheep(Animal):
    def sound(self):
        return "Baa"

# Create instances of Cow and Sheep
cow = Cow()
sheep = Sheep()

# Call the sound method on both instances
print(f"The cow says: {cow.sound()}")  # Output: The cow says: Moo
print(f"The sheep says: {sheep.sound()}")  # Output: The sheep says: Baa
```
